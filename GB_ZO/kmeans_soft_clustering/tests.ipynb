```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Define the number of samples and clusters
n_samples = 300
n_clusters = 10 # You can change this value

# Generate 2D blob data
X, y = make_blobs(n_samples=n_samples, centers=n_clusters, n_features=2, random_state=42, cluster_std=2.)

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50, alpha=0.7)
plt.title(f'{n_clusters} Clusters of 2D Data from make_blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```
